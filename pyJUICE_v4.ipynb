```python
#@title TRAIN A NEW VOICE

Name = "" #@param {type:"string"}
#@markdown - 10x WAV files, between 6-10 seconds each.

# I recommend you upload at least 2 audio clips. They must be a WAV file, 6-10 seconds long.
CUSTOM_VOICE_NAME = "(Name)"

import os
from google.colab import files

custom_voice_folder = f"tortoise/voices/{CUSTOM_VOICE_NAME}"
os.makedirs(custom_voice_folder)
for i, file_data in enumerate(files.upload().values()):
  with open(os.path.join(custom_voice_folder, f'{i}.wav'), 'wb') as f:
    f.write(file_data)
```

# ⚙️ Set Quality

```python
preset = "high_quality"  #@param ['ultra_fast', 'fast', 'standard', 'high_quality']
```

# 🤵 Single Sentance

```python
# Text to be Spoken
text= """
All vocal performances in this audio presentation 
are generated completely with stable diffusion machine learning.
"""

# Add an Inflection
inflection= """
(he said with joy)
"""

# Pick one of the installed voices
voice = 'sns'


# Load it and send it through Tortoise.
voice_samples, conditioning_latents = load_voice(voice)
gen = tts.tts_with_preset(text + inflection, voice_samples=voice_samples, conditioning_latents=conditioning_latents, 
                          preset=preset)
torchaudio.save('sample.wav', gen.squeeze(0).cpu(), 24000)
IPython.display.Audio('sample.wav')
```
